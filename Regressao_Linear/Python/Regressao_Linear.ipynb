```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

Importao dos dados com pandas

```python
upload = files.upload()
base = pd.read_csv('slr12.csv', sep=';')
base.shape
```

Vizualizao

```python
base.head()
```

Separando Variaveis dependentes e independentes

```python
X = base.iloc[:, 1].values
y = base.iloc[:, 0].values
X
```

Verificando corelao

```python
correlacao = np.corrcoef(X, y)
correlacao
```

Criando e Treinando modelo e ajustando X para o modelo

```python
X = X.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(X, y)
```

Gerando o Grafico

```python
plt.scatter(X, y)
plt.title('Taxa de franquia por Custo inicial')
plt.ylabel('Custo inicial')
plt.xlabel('Taxa de franquia')
plt.plot(X, modelo.predict(X), color = 'red', )
```

Prevendo valores

```python
modelo.predict([[1200]])
```

Visualizar os residuais

```python
visualizador = ResidualsPlot(modelo)
visualizador.fit(X, y)
visualizador.poof()
```
